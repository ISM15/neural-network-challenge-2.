## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]

```

```python
# Create a list of at least 10 column names to use as X data
x_cols = ['Age', 'BusinessTravel','DistanceFromHome', 'Education', 'EducationField','EnvironmentSatisfaction','HourlyRate', 'Job' 'Involvement', 'WorkLifeBalance']



# Create X_df using your selected columns
X_df = attrition_df[x_cols]


# Show the data types for X_df
X_df.dtypes

```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_df = pd.get_dummies(X_df)
X_df.head()

# y_df_encoded = pd.get_dummies(y_df)
# y_df_encoded.head()
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
y_df['Attrition'] = y_df['Attrition'].replace({'Yes': 1, 'No': 0})

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df)

```

```python
y_train.head()
```

```python
y_train.head()
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder


# Fit the encoder to the training data
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore', drop=None)
encoder.fit(np.array(y_train['Department']).reshape(-1, 1))



# Create two new variables by applying the encoder
# to the training and testing data
# X_train_encoded = encoder.fit_transform(X_train)
# X_test_encoded = encoder.fit_transform(X_test)
y_departnemnt_train_encoded = encoder.transform(np.array(y_train['Department']).reshape(-1, 1))
y_department_test_encoded = encoder.transform(np.array(y_test['Department']).reshape(-1, 1))


```

```python
# Create a StandardScaler
scaler = StandardScaler(with_mean=False)


# Fit the StandardScaler to the training data
X_scaler = scaler.fit(X_train)



# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
X_train_scaled.shape[1]


# Create the input layer
input_layer = layers.Input(shape=(X_train_scaled.shape[1],))


# Create at least two shared layers
shared_layer1 = layers.Dense(100, activation='relu')(input_layer)
shared_layer2 = layers.Dense(100, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer
department_branch = layers.Dense(100, activation='relu')(shared_layer2)
department_output = layers.Dense(3, activation='softmax', name='department_output')(department_branch)


# Create the hidden layer
attrition_branch = layers.Dense(100, activation='relu')(shared_layer2)


# Create the output layer
attrition_output = layers.Dense(1, activation='sigmoid')(attrition_branch)

```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer
attrition_branch = layers.Dense(100, activation='relu')(shared_layer2)
attrition_output = layers.Dense(1, activation='sigmoid')(attrition_branch)

# Create the hidden layer
attrition_branch = layers.Dense(100, activation='relu')(shared_layer2)


# Create the output layer
attrition_output = layers.Dense(1, activation='sigmoid', name='attrition_output')(attrition_branch)

```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[department_output, attrition_output])


# Compile the model
model.compile(loss={'department_output': 'categorical_crossentropy',
                    'attrition_output': 'binary_crossentropy'},
              optimizer='adam',
              metrics=['accuracy', 'accuracy'])


# Summarize the model
model.summary()
```

```python
y_departnemnt_train_encoded
```

```python
# Train the model

model.fit(X_train_scaled, {
    'department_output': y_departnemnt_train_encoded,
    'attrition_output': y_train['Attrition']
}, epochs=100, batch_size=32)



```

```python
# Evaluate the model with the testing data
results = model.evaluate(np.array(X_test_scaled), {
        'department_output': y_department_test_encoded,
        'attrition_output': y_test['Attrition']
    })
```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy: {results[1]}")
print(f"Attrition Accuracy: {results[2]}")
```
